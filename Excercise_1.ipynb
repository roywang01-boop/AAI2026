```python
import json

def mock_llm(prompt):
    """
    Simulates LLM behavior based on keywords in the prompt.
    This allows us to demonstrate prompt chaining without API calls.
    """

    if "triage assistant" in prompt:
        return json.dumps({
            "issue_types": ["billing_dispute", "delivery_problem"],
            "priority": "high",
            "order_id": "A1042",
            "missing_info": ["confirmation of shipping address", "screenshot of duplicate charge"]
        }, indent=2)

    elif "Ask ONLY for the missing information" in prompt:
        return json.dumps({
            "questions": [
                "Can you confirm the shipping address used for this order?",
                "Could you provide a screenshot of the duplicate charge?"
            ]
        }, indent=2)

    elif "senior customer support agent" in prompt:
        return json.dumps({
            "proposed_actions": [
                "Investigate duplicate transaction in billing system",
                "Initiate refund if duplicate confirmed",
                "Open delivery trace with carrier"
            ],
            "refund_policy_assumptions": [
                "Refund allowed for confirmed duplicate charges",
                "Refund allowed for lost packages after investigation"
            ],
            "customer_steps": [
                "Confirm address",
                "Provide screenshot"
            ],
            "internal_steps": [
                "Check payment logs",
                "Contact shipping carrier"
            ]
        }, indent=2)

    elif "support team lead" in prompt:
        return json.dumps({
            "escalate": True,
            "reason": "Duplicate charge and delivery marked delivered but not received",
            "final_customer_reply": "We sincerely apologize for the inconvenience. We are investigating the duplicate charge and the delivery issue. Once verified, we will process your refund promptly and keep you updated."
        }, indent=2)

    return "{}"
```

```python
customer_message = """
Hi, I was charged twice for order #A1042.
I only placed one order.
Also, the tracking says delivered but I never received it.
I need this fixed urgently.
"""
```

```python
prompt_1 = f"""
You are a customer support triage assistant.

Classify the issue and extract structured data.

Customer message:
{customer_message}
"""

output_1 = mock_llm(prompt_1)
print("Step 1 Output:")
print(output_1)
```

```python
prompt_2 = f"""
You are a support agent.

Based on this triage data:
{output_1}

Ask ONLY for the missing information.
"""

output_2 = mock_llm(prompt_2)
print("\nStep 2 Output:")
print(output_2)
```

```python
prompt_3 = f"""
You are a senior customer support agent.

Customer message:
{customer_message}

Triage data:
{output_1}

Provide structured resolution plan.
"""

output_3 = mock_llm(prompt_3)
print("\nStep 3 Output:")
print(output_3)
```

```python
prompt_4 = f"""
You are a support team lead.

Given:
Triage data:
{output_1}

Resolution plan:
{output_3}

Decide whether to escalate and generate final reply.
"""

output_4 = mock_llm(prompt_4)
print("\nStep 4 Output:")
print(output_4)
```
