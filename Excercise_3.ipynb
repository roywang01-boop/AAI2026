```python
original_text = """
Customer reported a double charge for order A1042 and says the tracking shows delivered
but the package was not received. Customer wants a refund urgently.
Support should ask for details and then fix it.
"""

bad_summary = "The customer has a problem with an order and needs help. They were charged twice and the delivery is wrong."
print("Bad summary:\n", bad_summary)
```

```python
self_reflection_prompt = f"""
You are a writing critic and editor.

Task:
A) Critique the summary using this rubric:
- Accuracy: includes all key facts and does not invent details
- Clarity: specific and unambiguous
- Completeness: captures billing + delivery issues
- Tone: professional and neutral
- Length: 35 to 60 words

B) Rewrite the summary to meet every requirement.

Text to summarize:
{original_text}

Summary to critique:
{bad_summary}

Output format:
1) Critique: bullet list
2) Improved summary: single paragraph
"""
print(self_reflection_prompt)
```

```python
ai_critique_and_revision = """
1) Critique:
- Too vague: it does not mention the order ID or the tracking marked delivered.
- Missing key detail: it does not clearly separate billing dispute vs delivery issue.
- Tone is okay, but it lacks actionable clarity.
- Length is short but sacrifices important facts.

2) Improved summary:
Customer reports a duplicate charge for order #A1042 and requests a refund. Tracking indicates the order was delivered, but the customer states the package was not received. Issue involves both a billing dispute and a potential delivery exception, requiring charge verification and shipment investigation before final resolution.
"""
print(ai_critique_and_revision)
```
